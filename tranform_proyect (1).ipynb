```python
import pandas as pd
df = pd.read_csv("/Traffic_Crashes.csv", encoding="latin1")
df = pd.DataFrame(df)
df
```

```python
columns_to_drop = [
    'LIGHTING_CONDITION', 'FIRST_CRASH_TYPE', 'TRAFFICWAY_TYPE', 'LANE_CNT',
    'ALIGNMENT', 'ROADWAY_SURFACE_COND', 'ROAD_DEFECT', 'REPORT_TYPE',
    'INTERSECTION_RELATED_I', 'NOT_RIGHT_OF_WAY_I', 'HIT_AND_RUN_I',
    'DATE_POLICE_NOTIFIED', 'PRIM_CONTRIBUTORY_CAUSE', 'SEC_CONTRIBUTORY_CAUSE',
    'BEAT_OF_OCCURRENCE', 'PHOTOS_TAKEN_I', 'STATEMENTS_TAKEN_I', 'DOORING_I',
    'WORK_ZONE_I', 'WORK_ZONE_TYPE', 'WORKERS_PRESENT_I', 'NUM_UNITS','CRASH_RECORD_ID','RD_NO',
    'CRASH_DATE_EST_I','TRAFFIC_CONTROL_DEVICE','DEVICE_CONDITION','DAMAGE','STREET_DIRECTION','MOST_SEVERE_INJURY'
]

df = df.drop(columns=columns_to_drop)
```

```python
df
```

eliminar valores nulos

```python
# Eliminar filas con valores nulos
df = df.dropna()
```

```python
df
```

convertir a enteros



```python
columnas_enteros = ['INJURIES_TOTAL', 'INJURIES_FATAL', 'INJURIES_INCAPACITATING',
                    'INJURIES_NON_INCAPACITATING', 'INJURIES_REPORTED_NOT_EVIDENT',
                    'INJURIES_NO_INDICATION', 'INJURIES_UNKNOWN', 'CRASH_HOUR','CRASH_DAY_OF_WEEK','STREET_NO','CRASH_MONTH']

# Convertir las columnas a enteros
df[columnas_enteros] = df[columnas_enteros].astype(int)
```

```python
df
```

elimina las palabras

```python
# para eliminar lo que va despues del/
df['CRASH_TYPE'] = df['CRASH_TYPE'].str.split('/').str[0].str.strip()

#para verificar bien
counted_crash_types = df['CRASH_TYPE'].value_counts()

print(counted_crash_types)
```

```python
#quitar la palabra and
df['CRASH_TYPE'] = df['CRASH_TYPE'].str.replace('AND', '').str.strip()

# Pa verificar como quedo
counted_crash_types = df['CRASH_TYPE'].value_counts()

print(counted_crash_types)
```

pasar de numeros a dias

```python
# Diccionario de mapeo de números a nombres de días
dias_semana = {
    1: 'Lunes',
    2: 'Martes',
    3: 'Miércoles',
    4: 'Jueves',
    5: 'Viernes',
    6: 'Sábado',
    7: 'Domingo'
}


df['CRASH_DAY_OF_WEEK'] = df['CRASH_DAY_OF_WEEK'].map(dias_semana)

```

acomodar en los meses

```python
# acomodar los meses
meses = {
    1: 'Enero',
    2: 'Febrero',
    3: 'Marzo',
    4: 'Abril',
    5: 'Mayo',
    6: 'Junio',
    7: 'Julio',
    8: 'Agosto',
    9: 'Septiembre',
    10: 'Octubre',
    11: 'Noviembre',
    12: 'Diciembre'
}

df['CRASH_MONTH'] = df['CRASH_MONTH'].map(meses)



```

```python
df
```

organizar coordenas

```python
# dejar solo las coordenadas y el orden que es
df['LOCATION'] = df['LOCATION'].str.replace(r'^POINT \((-?\d+\.\d+) (-?\d+\.\d+)\)', r'\2, \1')


```

```python
df
```

```python
nombre_archivo = 'Traffic_convertido.csv'

# Utiliza el método to_csv para guardar el DataFrame en un archivo CSV
df.to_csv(nombre_archivo, index=False)
```
